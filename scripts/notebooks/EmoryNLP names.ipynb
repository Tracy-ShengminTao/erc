```python
from glob import glob
import json

def get_name(path):
    with open(path, 'r') as stream:
        foo = json.load(stream)
    
    return foo['Speaker']

paths = glob('Datasets/EmoryNLP/raw-texts/*/*.json')
names = [get_name(path) for path in paths]
```

```python
from collections import Counter

names = dict(Counter(names))
names = [(key, val) for key, val in names.items()]
names.sort(key=lambda x:-x[1])
```

```python
names
```

```python
def reduce_name(name):
    prefix = ['Dr.', 'Mr.', 'A', 'The', 'Ms.', 'Mrs.']
    
    if name.split(' ')[0].lower() in [pre.lower() for pre in prefix]:
        name = ' '.join(name.split(' ')[:2])
    else:
        name = name.split(' ')[0]
    name = name.strip()
    return name
```

```python
names = [(reduce_name(name), count) for name, count in names]
```
