```python
from tqdm.notebook import tqdm
from glob import glob
import os
import pickle

for DATASET in tqdm(['AFEW', 'CAER', 'IEMOCAP', 'MELD']):
    for DS in tqdm(['train', 'val', 'test']):
        os.makedirs(os.path.join(DATASET, 'faces_', DS), exist_ok=True)
        faces_path = glob(os.path.join(DATASET, 'faces', DS, '*.pkl' ))
        for pickle_path in tqdm(faces_path):
            try:
                with open(pickle_path, 'rb') as stream:
                    foo = pickle.load(stream)
            except Exception as e:
                print(e)
                continue
                
            bar = {}
            for frame_idx, list_of_features in foo.items():
                bar[frame_idx] = []
                for features in list_of_features:
                    feature_dict = {'age': features.age,
                                    'bbox': features.bbox,
                                   'det_score': features.det_score,
                                   'gender': features.gender,
                                    'landmark': features.landmark,
                                    'normed_embedding': features.normed_embedding
                                   }
                    bar[frame_idx].append(feature_dict)
            with open(os.path.join(DATASET, 'faces_', DS, os.path.basename(pickle_path)), 'wb') as stream:
                pickle.dump(bar, stream)
```
